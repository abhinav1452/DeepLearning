```
import shutil
```

```
from tqdm import notebook
```

```
for image_name in notebook.tqdm_notebook(os.listdir('train')):
  if 'cat' in image_name:
    shutil.copy('train/'+image_name, 'master_data/cat')
  else:
    shutil.copy('train/'+image_name, 'master_data/dog')
```

```
import tensorflow as tf
```

```
idg = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255,validation_split=0.2)
```

```
train_idg = idg.flow_from_directory('master_data',target_size=(150,150),batch_size=64,subset='training')
val_idg = idg.flow_from_directory('master_data',target_size=(150,150),batch_size=64,subset='validation')
```

```
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=(150,150,3)))
model.add(tf.keras.layers.Dense(512,tf.keras.activations.relu))
model.add(tf.keras.layers.Dense(64,tf.keras.activations.relu))
model.add(tf.keras.layers.Dense(2,tf.keras.activations.softmax))
```

```
model.summary()
```

```
model.compile(tf.keras.optimizers.SGD(),loss=tf.keras.losses.binary_crossentropy,metrics=['acc'])
```

```
model.fit_generator(train_idg,epochs=10,validation_data=val_idg)
```
