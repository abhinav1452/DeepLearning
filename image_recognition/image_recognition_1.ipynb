```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
```

```
(x_train,y_train),(x_test,y_test) = keras.datasets.cifar10.load_data()
```

```
x_train.shape
```

```
y_train
```

```
 x_train_scaled = x_train/255
 x_test_scaled = x_test/255
```

```
y_train_catrgorical = keras.utils.to_categorical(y_train, num_classes=10,dtype='float32')
y_test_catrgorical = keras.utils.to_categorical(y_test, num_classes=10,dtype='float32')
```

```
y_train_catrgorical[:5]
```

```
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (32,32,3)),
    keras.layers.Dense(3000, activation='relu'),
    keras.layers.Dense(1000, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])
```

```
model.summary()
```

```
model.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train_scaled,y_train_catrgorical,epochs=50)
```

```
model.predict(x_test_scaled)
```

```
model.evaluate(x_test_scaled,y_test_catrgorical)
```
